#### 삼성전자 주가 차트 (기본 candle chart)

```python


stock_name = '삼성전자'
fig = go.Figure(data = [go.Candlestick(x=df['날짜'],
                                       open=df['시가'],
                                       high=df['최고가'],
                                       low=df['최저가'],
                                       close=df['종가'])])


fig.layout = dict(title=stock_name, 
                       xaxis = dict(type="category", 
                                    categoryorder='category ascending'))
fig.update_xaxes(nticks=5)


fig.show()


```
